```python
#computer vision : technology 

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((200,500,3),np.uint8)
#(h,w)--->this is hwo open cv takes it
img
```

```python
plt.imshow(img)
```

```python
# Draw circle

circle = cv2.circle(img,(250,100), 50, (255,0,0),-1)
      #img,center,radius,colour,-1 to fill the circle 
plt.imshow(circle)
```

```python

# Rectangle 

rec = cv2.rectangle(img,(150,150),(50,50),(255,255,0),3)
 #----> img,pt1,pt2,colour,thickness
plt.imshow(rec)
```

```python
# Drawing line

l = cv2.line(img,(50,50),(150,150),(0,0,255),2)
# pt1,pt2
plt.imshow(l)
     
```

```python

text = cv2.putText(img,'Hey',(350,120),cv2.FONT_HERSHEY_COMPLEX_SMALL,2,(255,255,255),2)
plt.imshow(text)
     
```

FACE DETECTION 

```python

img = cv2.imread('/content/face.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
haarCas = '/content/haarcascade_frontalface_alt (1).xml'
haarCas
     
```

```python
#USE CASCADE CLASSIFIER 
faces = cv2.CascadeClassifier(haarCas)
faces
     
```

```python
face = faces.detectMultiScale(img,scaleFactor=1.1,minNeighbors=4)
face
```

```python

for (x,y,w,h) in face:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)

plt.imshow(img)
```

NUMBER PLATE DETECTION 

```python
img = cv2.imread('/content/IMAGE.JPG')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
     
```

```python
haarCas = '/haarcascade_russian_plate_number (1).xml'
haarCas
```

```python
NumberP = cv2.CascadeClassifier(haarCas)
NumberP
```

```python

npp = NumberP.detectMultiScale(img,scaleFactor=1.1,minNeighbors=4)
npp
```

```python
for (x,y,w,h) in npp:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),7)

plt.imshow(img)
```
